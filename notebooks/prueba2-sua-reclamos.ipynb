```python
import numpy as np
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
import sys
import os.path
sys.path.append('../modules')
from geomr import location_rosario, projmr_to_latlon
```

```python
df_acc = pd.read_excel('../fuentes/exportacionTipo-reclamo y Fecha agosto 2021.ods', engine='odf')
```

```python
df_acc.head()
```

```python
t_list = ["Stamen Terrain", "Stamen Toner", "Mapbox Bright", 'cartodbpositron']
```

```python
map = folium.Map(location=location_rosario(), tiles = "cartodbpositron", zoom_start = 12)
```

```python
heat_data = [list(projmr_to_latlon(rows["Coord. X"], rows["Coord. Y"])) for index, rows in df_acc.iterrows()]
```

```python
plugins.HeatMap(heat_data, radius=4, blur=8).add_to(map)
```
