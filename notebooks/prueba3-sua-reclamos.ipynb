```python
import numpy as np
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
import sys
import os.path
sys.path.append('../modules')
from geomr import location_rosario, projmr_to_latlon
from pandasmr import clasifico_por_dia
from proxymr import set_proxy
set_proxy()
```

```python
df_acc = pd.read_excel('../fuentes/exportacionTipo-reclamo y Fecha agosto 2021-short1.ods', engine='odf')
```

```python
map = folium.Map(location=location_rosario(), tiles = "cartodbpositron", zoom_start = 12)
```

```python
resultado = clasifico_por_dia(df_acc)
```

```python
hm = plugins.HeatMapWithTime(data=list(resultado.values()),
                             index=list(resultado.keys()), 
                             radius=10,
                             auto_play=True,
                             max_opacity=0.3)
```

```python
hm.add_to(map)
```
